```python
import nltk
import re

nltk.download('punkt')

def count_syllables(word):
    # Simple syllable counting based on vowels
    vowels = "aeiouy"
    count = 0
    word = word.lower()

    # Remove trailing 'e' and handle words ending with "es" or "ed"
    if word.endswith('e'):
        word = word[:-1]
    if word.endswith(('es', 'ed')):
        return count

    for char in word:
        if char in vowels:
            count += 1

    return max(1, count)
```

```python
import pandas as pd
from textblob import TextBlob

# Assuming 'df2' is your DataFrame and 'output' is your existing Excel file

# Iterate over the rows in df2
# for index, row in df2.iterrows():
#     article_text = str(row['article text'])  # Convert to string
#     if pd.notna(article_text):  # Check for NaN
#         # Perform sentiment analysis using TextBlob
#         blob = TextBlob(article_text)
#         positive_score = sum(1 for sentence in blob.sentences if sentence.sentiment.polarity > 0)
#         negative_score = sum(1 for sentence in blob.sentences if sentence.sentiment.polarity < 0)
#         polarity_score =(positive_score - negative_score)/((positive_score + negative_score) + 0.000001)

#         # Update the specific row in the output DataFrame with the scores
#         output.at[index, 'POSITIVE SCORE'] = positive_score
#         output.at[index, 'NEGATIVE SCORE'] = negative_score
#         output.at[index, 'POLARITY SCORE'] = polarity_score

# Save the updated DataFrame to the same Excel file
# output.to_excel('/content/Output Data Structure (1).xlsx', index=False)

from textblob import TextBlob

# Assuming 'df2' is your DataFrame
for index, row in df2.iterrows():
    article_text = str(row['article text'])

    # Perform sentiment analysis using TextBlob
    blob = TextBlob(article_text)
    positive_score = sum(1 for sentence in blob.sentences if sentence.sentiment.polarity > 0)
    negative_score = sum(1 for sentence in blob.sentences if sentence.sentiment.polarity < 0)

    # Calculate Polarity Score
    total_score = positive_score + negative_score
    polarity_score = (positive_score - negative_score) / (total_score + 0.000001)
    # Calculate Subjectivity Score
    total_words = len(article_text.split())  # Total words after cleaning
    subjectivity_score = (positive_score + negative_score) / (total_words + 0.000001)
    # Calculate Average Sentence Length
    sentences = article_text.split('.')  # Assuming sentences end with a period (.)
    num_words = len(article_text.split())
    num_sentences = len(sentences)
    avg_sentence_length = num_words / (num_sentences + 0.000001)

    # Calculate Percentage of Complex Words
    complex_words = [word for word in article_text.split() if len(word) > 2]  # Assuming words with more than 2 characters are complex
    num_complex_words = len(complex_words)
    percentage_complex_words = num_complex_words / (num_words + 0.000001)

    # Calculate Fog Index
    fog_index = 0.4 * (avg_sentence_length + percentage_complex_words)

   # Calculate Average Number of Words Per Sentence
    sentences = article_text.split('.')  # Assuming sentences end with a period (.)
    num_words = len(article_text.split())
    num_sentences = len(sentences)
    avg_words_per_sentence = num_words / (num_sentences + 0.000001)
    # Calculate complex_word_count
    words = article_text.split()
    complex_word_count = sum(1 for word in words if textstat.syllable_count(word) > 2)
    # Calculate Word Count
    words = nltk.word_tokenize(article_text)
    # Remove stop words and punctuation
    stop_words = set(stopwords.words('english'))
    words = [word.lower() for word in words if word.isalpha() and word.lower() not in stop_words]
    word_count = len(words)
    words = nltk.word_tokenize(article_text)

    # Calculate Syllable Count Per Word
    syllable_count_per_word = sum(count_syllables(word) for word in words) / len(words)

    # Calculate Personal Pronouns count
    personal_pronouns = re.findall(r'\b(?:I|we|my|ours|us)\b', article_text, flags=re.IGNORECASE)
    personal_pronouns_count = len(personal_pronouns)

    # Calculate Average Word Length
    average_word_length = sum(len(word) for word in words) / len(words)

    # Update the DataFrame with the score
    df2.at[index, 'WORD COUNT'] = word_count
    df2.at[index, 'POSITIVE SCORE'] = positive_score
    df2.at[index, 'NEGATIVE SCORE'] = negative_score
    df2.at[index, 'POLARITY SCORE'] = polarity_score
    df2.at[index, 'SUBJECTIVITY SCORE'] = subjectivity_score
    df2.at[index, 'AVG SENTENCE LENGTH'] = avg_sentence_length
    df2.at[index, 'PERCENTAGE OF COMPLEX WORDS'] = percentage_complex_words
    df2.at[index, 'FOG INDEX'] = fog_index
    df2.at[index, 'AVG NUMBER OF WORDS PER SENTENCE'] = avg_words_per_sentence
    df2.at[index, 'COMPLEX WORD COUNT'] = complex_word_count
    df2.at[index, 'WORD COUNT'] = word_count
    df2.at[index, 'SYLLABLE PER WORD'] = syllable_count_per_word
    df2.at[index, 'PERSONAL PRONOUNS'] = personal_pronouns_count
    df2.at[index, 'AVG WORD LENGTH'] = average_word_length
```
